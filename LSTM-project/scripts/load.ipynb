```python
import pandas as pd
```

```python
data202012 = pd.read_csv("한국전력공사_일자별_충전방식별_시간대별_충전부하_2020_2022.csv", encoding='euc-kr')
# data2021 = pd.read_excel("한국전력공사_EV시간대별 충전부하_202101_202206.xlsx", encoding='euc-kr')
data2023 = pd.read_csv("한국전력공사_전기차 시간대별 충전부하_20231231.csv", encoding='UTF-8')
data2024 = pd.read_csv("한국전력공사_전기차 시간대별 충전부하_20240930.csv", encoding='UTF-8')
```

```python
# data2024 컬럼명 통일 ('01시' -> '1시', '02시' -> '2시' 등)
data2024_renamed = data2024.copy()
data2024_renamed.columns = data2024_renamed.columns.str.replace(r'^0(\d)시$', r'\1시', regex=True)
data2024_renamed.columns = data2024_renamed.columns.str.replace('00시', '0시')

# 2020~2024년 데이터를 하나로 합치기
data = pd.concat([data202012, data2023, data2024_renamed], ignore_index=True)
```

```python
# 통합 데이터 csv로 저장
data.to_csv('dataset.csv', index=False, encoding='utf-8-sig')
```
